## 01)Definitions

__What is a Data-Set?__
- A data set is a collection of discrete samples of related data that may be accessed individually or in combination or managed as a whole entity.

__What is EDA?__
- It is Data Exploration technique to undrestand several aspects of data.
- Undrestand the data and be able to interpret it, in order to do more complecated process.
- Find the relationship among variables,which give us a wider perspective on data.
- Find Null values and NAN(Not a Number) values, and replace them using a correct value.
