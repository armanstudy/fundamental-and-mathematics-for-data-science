## 01)Uniform distribution

- In order to distinguish the data distribution we hav used histogram until now.

- To do that we use QQPLOT from `stats.api` .
- QQPLOT check the dispersion and skewness. 

## 02) QQplt in python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.api import qqplot,qqplot_2samples
```

```python
rnd_normal = np.random.normal(size = 10_000)
sns.histplot(rnd_normal, kde=True)
```

- Using histogram we can undrestand that it is a normal distribution, but is it a standard normal distribution? let's use qqplot.

```python
qqplot(rnd_normal, line='45');
```

- the line means normal distribution.
- The closer the chart is to 45 degrees, the closer it is to the standard normal distribution. Here our distribution is completely normal.The closer the chart is to 45 degrees, the closer it is to the normal distribution. Here our distribution is completely normal.

- Let's define a random distribution with std equal to 5.

```python
rand_normal = np.random.normal(size = 10_000, scale=5)
sns.histplot(rand_normal, kde=True)
```

- __Check the similarity with standard distribution.__

```python
qqplot(rand_normal, line='45');
```

    - the distribution is normal but not standard.

__example__

- Check a uniform dustribution using qqplot:

```python
rand_uniform = np.random.uniform(-3, 3, size=10_000)
```

```python
qqplot(rand_uniform, line='45');
```

- The whole distribution is not normal, but between -1 to 1 it is normal, since it is a line and overlap with normal distribution as below:

```python
plt.hist(rnd_normal, bins=30)
plt.hist(rand_uniform, bins=30, alpha=0.6)
plt.show()
```

## 03)Comparing two distributions together

```python
qqplot_2samples(rnd_normal, rand_uniform, line='45');
```

## 04)qqplot on data set

- in student performance data set, we will raise some grades and comare with the original ones using qqplot_2samples.

```python
data = pd.read_csv("G:\\Data_analyst\\student_data.csv")
```

```python
grade = data['G1'] + data["G2"] + data["G3"]
```

```python
grade_new = grade.copy()
```

```python
grade_new[grade_new < 43] = grade_new[grade_new < 43] + grade.std()
grade_new[grade_new > 43] = grade_new[grade_new > 43] + (grade.std())/4
```

-first we check deviation relative to normal.

```python
qqplot(grade, line='45');
```

- It is normal to some extent but not standard normal.

- Check the differnce between grade and grade_new.

```python
qqplot_2samples(grade, grade_new , line='45');
```
