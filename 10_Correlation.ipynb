## 01)Definition

- Correlation measures the relationship between two variables.

## 02)Correlation Coefficient

-  1 = there is a perfect linear relationship between the variables.
-  0 = there is no linear relationship between the variables.
- -1 = there is a perfect negative linear relationship between the variables 

## 03)correlation using scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

**correlatin for G1 and G2**

```python
data = pd.read_csv("G:\\Data_analyst\\student_data.csv")
```

```python
plt.scatter(data['G1'],data['G2'])
plt.show()
```

```python
from scipy.stats import pearsonr
cov, _ = pearsonr(data['G1'], data['G2'])
cov
```

- We can see that the two variables are positively correlated and that the correlation is 0.8. This suggests a high level of correlation, e.g. a value above 0.5 and close to 1.0.

## 04)correlation for whole dataset

```python
plt.figure(figsize=(12, 12), dpi = 120)
sns.heatmap(data.corr(), annot=True, cmap='CMRmap')
plt.show()
```
