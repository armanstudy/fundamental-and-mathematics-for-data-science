## 1) statistics and probability

- When we want to gain an understanding of a data set, we need to know `statistics and probability` in order to describe the behavior of the data.
- Statistics literally means counting.
- Statistics is a science that collects, summarizes, classifies, describes and analyzes data.
- Summarizing, classifying, describing and analyzing data is used to predict and estimate events. 
- The probability is the measure of the likelihood of an event to happen. 

- We always express opinion about what's going on around us and say the chances of its happening is high or low. But it is qualitative and doesn't have a number. We want to calculate this chance of happening and get a number for this. This method is called probability. So in probability, we want to turn the chances of something happening into numbers.

## 2)Sample Space

- All possible situations in an experiment.
- For example, in rolling a dice, set of  all possible outcomes **samples** is equal to:
- `S = {1, 2, 3, 4, 5, 6}`
- Or in coin_flipping the Sample Space would be:
- `S = {Head, Tail}`

##  3)Event space

- In an experiment with sample space  S , an event space is any collection of (some, all, or none) outcomes. Or we can say event is any subset of S.
- For instance, if we flip a coin one time and T (tail) comes as outcome, then {T} is an event, a subset of S = {H, T}.

## 4)Calculate the probability of an event

- The probability of an event space is equal to the division of `number of event members` by `number of sample space members`.
- For example, if we want to calculate the probability of occurrence {2, 3} in rolling a dice for 2 times, {2, 3} is the our Event space, E = {2, 3}, which has 2 members. And our sample apace is, S = {1, 2, 3, 4, 5, 6}, which has 6 members, then as the above definition says, the probability of our E would be : `number of event members` by `number of sample space members`:

### $$Probability of E = \frac{number of event members}{number of sample space members} = \frac{2}{6}=\frac{1}{3}$$

- Now we can say the probability of outcomes {2, 3} in rolling a dice for 2 times is equal to :
## $$\frac{1}{3}$$


- The mathematical formula for calculating the probability is shown like below : 

## $$P(E) = \frac{n(E)}{n(s)}$$
    

n(E) is the number of E_members, the events which we want to calculate thier probability of happening. 

n(s) is the number of S_members.

The probability value is always :   $0\le P(E) \le1$

Zero means impossible and one means certainty. But in nature, we can never say with certainty that the probability of something happening is zero or one

__Example: calculate the probability for occuring 1 in rolling a dice for one time.__

- P(1) = ?
- S = {1, 2, 3, 4, 5, 6} $\rightarrow$(S) = 6
- E = {1} $\rightarrow$ n(E) = 1
$$P(1) = \frac{n(E)}{n(S)} = \frac{1}{6}$$

__Example : Calculate the probability for occuring  1 or 2 or 5 in rolling a dice.__

- E = {1, 2, 5} $\rightarrow$ n(E) = 3
- S = {1, 2, 3, 4, 5, 6} $\rightarrow$ n(s) = 6
$$P(\{1, 2, 5\}) = \frac{3}{6} = \frac{1}{2}$$

## 5)Conditional Probability

- Suppose event B has occurred with positive probability.
- We show the probability of event A under the condition B as `P(A|B)` .
- This probability is defined as follows : 
$$P(A|B) = \frac{P(A \cap B)}{P(B)} ; P(B)>0$$
- A : Event
- B : Condition


__Example:If the outcome of a dice roll is not equal to even number, find the probability that 1 will occur:__

- Sample Space : S = {1, 2, 3, 4, 5, 6} $\rightarrow$n(S) = 6
- Condition : B = {1, 3, 5}
- Event : A = {1} $\rightarrow$n(E) = 1
 $$P(A \cap B) = P(\{1\} \cap \{1, 3, 5\}) = P(\{1\}) = \frac{n(E)}{n(s)} = \frac{1}{6}$$
 $$P(B) = \frac{3}{6} = \frac{1}{2}$$
 $$P(A|B) = \frac{P(A \cap B)}{P(B)}=  \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}$$

## 8)Kolmogorov's three principles

__1)__ For any event A, the probability of A is greater or equal to 0 :
$$P(A) \ge 0$$

__2)__

$$P(Sample Space) = 1$$

- From the above formula we can conclude:

$$A \cup A' = S$$
$$A \cap A' = \emptyset$$
$$P(A)+P(B)=1$$

__3)__ If two sets are incompatible ($A \cap B = \emptyset$):
$$P(A \cup B) = P(A) + P(B)$$

__Example:  In rolling a dice, calculate the probability for occuring {1, 2} and {3} :__

- E(A) = {1, 2}
- E(B) = {3}
- Since A and B are incompatible :
$$P(A \cup B) = P(A) + P(B)$$

$$P(A) = \frac{1}{3}$$
$$P(B) = \frac{1}{6}$$
$$P(A \cup B) = \frac{1}{3} + \frac{1}{6} = \frac{1}{2}$$

__Example: In throwing a coin 2 times, calculate the probability for occuring at least one Head:__

- $S = \{(H,T),(T,H),(H,H),(T,T)\}$
- $E = \{(H,T), (T, H), (H,H) \}$
-$P(E) = \frac{3}{4}$

- On the basis of Kolmogorov's principles : 
- for n incompatible sets:
$$P(A \cup B\cup C ....) = P(A) + P(B) + P(C) + ...$$
$$P(\cup A_i ) = \sum\limits_{i=1}^{\infty}P(A_i)$$

- __1)Write a function that calculates the probability of the event by receiving the sample space and the event in rolling a dice.__

- Note 1: 
- our Sample Space is : S = {1, 2, 3, 4, 5, 6}
- Now if in our input Event we have repeated number, how should we omit it?
- suppose our E is : 
- E = {1, 2, 3, 3}

- In real world problems we determine that our input should be what type of data like List,Set,Dict,...
- In this situation, one way is to get the input(E) as a list an omit the repeated number by a simple loop.
- But the best way is to use Set data type for input,as the Set omit the repeated data.

- Note 2 :
- suppose our E as an input contains a number which is not in our S.
- To recognize this issue, we can use `set.issubset` method.
- For instance, 
- S = {1, 2, 3, 4, 5, 6}
- E = {1, 2, 7}

```python
S = {1, 2, 3, 4, 5, 6}
E = {1, 2, 3 , 7}
#we can take True or False using:
E.issubset(S)
```

- Now we should omit 7(repeated number) in order to calculate the Probability.
- In the below line we omit the repeated number and update the E simultaneously.

```python
E.intersection_update(S)
print(E)
```

### The final code

```python
def probability(S, E):
    if E.issubset(S):
        return round(len(E)/len(S),2)
    else:
        E.intersection_update(S)
        return round(len(E)/len(S),2)
```

```python
output_probability = probability({1, 2, 3, 4, 5, 6}, {1,2,3,10,2,5})
output_probability
```

- It means, for example if we rolled a dice 1000 time, the times that 1 has happened has been 100.So the probability of occuring 1 is 0.1  .And also for other numbers is the same, for 2 in 1000 times rolling dice the number of happening 2 is 200, so the probability is 0.2 `SO the sum of probabilities is not equal to 1` .

- To solve this problem lets use ditionary data type, since it omit the repeated data.

```python
space = {1 : 0.1, 2:0.2, 3:0.3, 4:0.4, 5:0.5, 6:0.6}
event_1 = {1, 2, 3, 1000, 50, 4}
```

```python
def probability2(space, event_1):
    s = set(space.keys())    #change to set datatype in order to be able to use issubset and intersection
    prob_event_1 = sum(space.values())
    p_sum = 0                    #To add the sum of probabilities of given event 
    if event_1.issubset(s):
        for e in event_1:
            p_sum += space[e]
        return round(p_sum/prob_event_1, 2)
    else:
        event_1.intersection_update(space)
        for e in event_1:
            p_sum += space[e]
        return round(p_sum/prob_event_1, 2)
```

```python
print(probability2(space, event_1))
```
