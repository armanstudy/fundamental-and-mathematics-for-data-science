## 05)Normal Distribution

- Normal distribution is exactly as bellow :

- In a standard normal distribution :
    $$mean \approx median \approx mode$$

- 68.2% of data has benn distributed between Mean-std  and mean+std.

__Test on a normal distributed array__

```python
arr = np.random.normal(size = 10000)
```

```python
std = arr.std()
std
```

```python
mean = arr.mean()
mean
```

```python
#mean-std
down = mean - std
down
```

```python
#mean+std
up = mean + std
up
```

```python
#calculate the number of data between mean+std and mean-std
arr_1 = arr[(arr>down) &  (arr < up)]
arr_1.size
```

- it means 6819 samples are between mean-std and  mean+std which is :
$$percrnt = \frac{6849}{10000} = 68.49 \% \approx 68.2\% $$

```python
sns.histplot(arr, bins = 20, kde = True)
```

## 06)Data distribution

- The center of normal distribution is not  zero and can have different centers:

```python
arr1 = np.random.normal(size = 10000, loc = 1, scale = 4)
arr2 = np.random.normal(size = 10000, loc = 5, scale = 10)
#loc is center or mean and scale is std 
```

```python
plt.hist(arr1, bins = 50)
plt.hist(arr2, bins = 50, alpha = .5)
plt.show()
```
