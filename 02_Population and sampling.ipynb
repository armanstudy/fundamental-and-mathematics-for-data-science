## 01) Definitions

- Statistical population and statistical sample is one of the primary topics in research. Researchers usually start their work by describing information (descriptive statistics) and try to generalize what they have obtained from examining the sample group to larger similar groups or statistical population (inferential statistics).

__Statistical Population:__

- A statistical population is a collection of all people, groups, objects or events that have one or more common characteristics. The number of members of the community is called the size of the community and is represented by the capital letter N.
- Example: the community of employees working in a bank

__Statistical sample:__

- A statistical sample is a smaller group of the population that is selected according to certain rules for observation and analysis and should be representative of the population. The results of a sample that is not representative of the society cannot be generalized to the society. 

- suppose we intend to evaluate weight parameter of a country with population of 80 million.But evaluting this population is difficult task.So we choose for instance 500k of population which are Representative of the whole community.

- Sampling has 4 steps and we are not allowed to choose the sample space randomly always.
- 1)choose the whole `population`.
- 2)We should choose our target population(For example, in evaluating the weight parameter of a country, if we are going to evaluate women weight, the women population would be our `target population`.
- 3)the member of target population must be `accessible`. For example, in weight evaluation, all women are not accessible for our experiment.Therefore, our population will becomes smaller.
- 3)Now we can choose our `sample space` from accessible population using methods which we talk about in the future.

- After evaluating on sample space we would be able to generalize the obtained rules to the whole population.

## 02)Data concept

- Anything that gives us information is data. such as image, video, text, table (weight, age).

__Statistical Population:__

- In most data analysis problems, we have a table to evaluate statistically.The table has rows and columns. Here our rows are the `samples`, such as peaple's name and columns are `variable(feature)` like weigth,height phone number and email.

- In Pandas we call this table a DataFrame. It could be an excel table that we call it in python using pandas.

__Kinds of variable:__

- The variable is divided into two categories:
1) `Numerical` such as age and income.
2) `Categorical` such as postal code.
- In order to distinguish them,remeber that in numerical variables we are able to compare them using greater and smaller, but in categorical we cannot use this and greater and smaller does not work in categorical variables.

- `Namerical` variables is divided into **Interval_scaled** and **Ratio_scaled** categories.
- **Interval_scaled** variable can be between $-\infty$ to +$\infty$, such as temprature.
- **Ratio_scaled** variable can be between 0 to +$\infty$ like the number of cars.
- **NOTE:** in ratio_scaled 0 means inexistency (when we say zero car,it means there is no car.),while in interval 0 shows an amount(the temprayure is 0C)  

- `categorical` variables is divided into **Ordinal** and **nominal** categories.
- In **Ordinal** variable the order is important such as education degree.
- In **Nominal** variable the order is not important such as postal code. 

- **Note:** we cannot use categorical variables in our machine learning model. Suppose our samples are some colors(red, green, yellow) these words does not have any meaning for machine, so we change them to numerical variables using `One_Hot_Encoding`.

## 03)Sampling methods

__Sampling__

- Sampling is a technique to select a subset of the population in order to make statistical inferences from them and estimate the characteristics of the entire population.

__kinds of sampling__

__1) Simple Random Sampling:__
- One of the best probability sampling techniques that help save time and resources is simple random sampling. It is a reliable method of obtaining information in which individual members of a population are selected randomly and purely randomly. Each individual has the same probability of being selected as part of a sample.

__2)Systematic Sampling:__
- In this sampling, the populathin is listed sequentioally and the samples will be selected at regular interval.

__3)Stratified Sampling:__
-  In stratified sampling, the population woulde be divided to subgroups, based on characteristics such as gender or age. Then, the samples will be choosen using another probability sampling method from each subgroup.

__4)Cluster Sampling:__
- In Cluster sampling the population would be divided to into clusters, such as city or school, and then randomly select some of these clusters as sample. The clusters be representations of the population.
