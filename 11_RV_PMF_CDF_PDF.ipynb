## 01)RV

- A random variable(RV) is a numerical description of the outcome of a statistical experiment.

- RV is of 2 kinds : 
- 1) Discrete : like roling a dice
- 2) continuous : like height, weight

## 02)PMF(Probability Mass Function)

- A probability mass function is a function over the sample space of a discrete random variable X which gives the probability that X is equal to a certain value. 
$$P_X(x) = p~(\{X=x\})$$

- for instance in roling a dice, the PMF for X=1 is:
-$$P_X(1) = p(\{ X=1 \}) = \frac{1}{^6}$$

## 03) CdF(Cumulative distribution function)

- The CDF of a real-valued random variable X, evaluated at x, is the probability function that X will take a value less than or equal to x. It is used to describe the probability distribution of random variables in a table.

$$F_X(x) = p~(\{X \le x\})$$

- For instance in roling a dice F(3) is:
$$F(x) = P(X \le 3) = P([1, 2, 3]) = \frac{1}{6} +\frac{1}{6} + \frac{1}{6} = \frac{1}{2} $$

-NOTE:
$$if(X\le0.5) \Longrightarrow F(x) = 0$$
$$if(x\le1.8) \Longrightarrow F(x) = p([1]) = \frac{1}{6}$$
$$F(7) = P(X \le 3) = P([1, 2, 3, 4, 5, 6]) = \frac{1}{6} +\frac{1}{6} + \frac{1}{6} +  \frac{1}{6}+  \frac{1}{6} +  \frac{1}{6} =1 $$
$$\Sigma P_X(x) = 1$$



## 04)Example in python

- we role a dice for 1000 times and then we will see the CDF charts:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dice = np.array([1, 2, 3, 4, 5, 6])
```

```python
# roling the dice for 1000 times:
rnd = np.random.choice(dice, size=1000)
```

```python
#Hisrogram:
plt.hist(rnd, bins=20)
plt.show()
```

```python
#Plot CDF:
plt.hist(rnd, cumulative=True)
plt.show()
```

```python
plt.hist(rnd, cumulative=True, density=True)
plt.show()
```

## 05) PDF(Probability Density Function)

- The Probability Density Function(PDF) defines the probability function representing the density of a continuous random variable lying between a specific range of values.

$$P(a \le X \le b) = (\int_{a}^{b} F_X(x)dx)$$
