## 01)3rd moment_skewness

- The normal distribution helps us to recognise skewness. When we talk about normal distribution, data has been distributed symetricall. The symmetrical distribution has no skewness and the tendency of data is to center.
$$mean \approx median \approx mode$$
- There is no skewness in normal distribution.The normal distribution has a skewness of 0.

- In a standard normal distribution the data has been devided to 2 50% around the mean.

### lest's check a normal distribution:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- Define a normal random distribution:

```python
arr = np.random.normal(size = 10_000, loc = 0)
```

```python
sns.histplot(arr, bins = 50, kde = True)
```

```python
arr_left_size = arr[arr<0]
arr_right_size = arr[arr>0]
```

- compare the left and right data size:

```python
arr_left_size.size
```

```python
arr_right_size.size
```

- It is clear that in a standard normal distribution:
    $$ DataLeftsize \approx DataRightsizesize $$

- But in skewness, data tendency is to the right or left.
- Skewness is of two types:
- **`Positive skewness`**: When the right of the distribution is longer,the data is positively skewed.
- **`Negative skewness`**: When the left tail of the distribution is longer, the distribution is negatively skewed. 

## 2)negative Skewness

- In negative skewness data tendency is to the right.

- In negative Skewness :
    $$mode > median > mean$$ 

## 03)positive Skewness

- In positive skewness :
$$mode < median < mean$$

## 04) 4th moment_Kurtosis

- Kurtosis is the characteristic of being flat or peaked. 

## 05)Formulas for moments

__1st Moment formula:__

$$center of data = \frac{(x-\overline{x})^1}{n} $$

- The 1st_moment comes from the power 1 in the formula.

__2nd Moment formula:__

$$Var = \frac{\Sigma(x-\overline{x})^2}{(n-1)}$$

- The 2nd_moment comes from the power 2 in the formula.

__3rd Moment formula:__

$$skew = \frac{\Sigma(x-\overline{x})^3}{(n-1)\times \sigma^3}$$

- The 3rd_moment comes from the power 3 in the formula.

__4th Moment formula:__

$$skew = \frac{\Sigma(x-\overline{x})^4}{(n-1)\times \sigma^4}$$

## 06) skewness and kurtosis on dataset

```python
data = pd.read_csv('G:/Data_analyst/student_data.csv')
df_grade = (data['G1'] + data['G2'] + data['G3'])
```

```python
sns.histplot(df_grade, bins = 50, kde = True)
```

- Our plot is like normal distribution.

```python
df_grade.mean()
```

```python
df_grade.median()
```

- mean and median are so near, the distribution might be normal.
- But we check the skewness and kurtosis.

```python
df_grade.skew()
```

- What does this number mean?

- **if**   **$-0.5 < skew < +0.5  \Longrightarrow  $**    **The distribution is normal** 

- **if**   $-1< skew < -0.5  ~~OR~~  +0.5 < skew < +1    \Longrightarrow  $   **the distribution is moderately skewed.** 

- **if** $skew < -1 ~and~ skew > +1  \Longrightarrow  $ **the distribution is highly skewed**

- Here the skewness is between -0.5 and 0.5 so the distribution is approximately symmetric or normal.

### calculate kurtosis:

```python
df_grade.kurtosis()
```

- When kurtosis is equal to 3, the distribution is `normal`.
- If the kurtosis is greater than 3 then it shows `positive kurtosis`.
- If the kurtosis is less than 3 then it shows `negative kurtosis`.

- Here `-0.332` is the difference with normal kurtosis number(3), which means the distribution has small kurtosis.
